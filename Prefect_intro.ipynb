# Introduction to Orchestation and Sequencing of Workflows with Python
Workflow orchestations refers to the process of defining, coordinating and automating the execution of multiple computational tasks that together form a larger process. Instead of running scripts independtly, orchestation structure these processes as pipelines, where each task is executed in a specific arder and may depend on the output of previous tasks.

In modern data processing systems, machine learning workflows, and infrastructure automation, processes are commonly divided into stages such as data acquisition, transformation, analysis, storage, and reporting. Workflow orchestration allows these stages to be modeled as a structured flow of tasks, enabling controlled execution and management of events such as failures, retries, and scheduled runs.

We can think of workflows as Directed Acylic graphs (DAG). Each node represents a task and the edges represent the dependence between them. This kind of structure allow us to clearly define which tastks we must run sequentially, and which ones in parallel. 

In the Python ecosystem, there are several frameworks that have been developed specifically for workflow orchestation, for example, Prefect and Apache Airflow. We can define the tasks, the dependencies, the schedule execution and monitor the behavior of complete pipelines.

## What code it should run
We begin with a little program that fetches a list of customers ids and process them in prallel

```python
from prefect import flow, task
import random 

@task
def get_customer_ids() -> list[str]:
    # Fetch customer IDs from a database or API
    return [f"Customer{n}" for n in random.choices(range(100), k=10)]

@task 
def process_customer(customer_id: str) -> str:
    # Process a single customer 
    return f"Processed {customer_id}"

@flow
def main() -> list[str]:
    customer_ids = get_customer_ids()
    # Map the process_customer task across all customer files
    results = process_customer.map(customer_ids)
    return results

if __name__ == "__main__":
    main()
```

## Where should it run?
We already have run locally a main() file, we now use use serve() to update the file

```python
from prefect import flow, task
import random

@task 
def get_customer_ids() -> list[str]:
    # Fetch customer IDs from a database or API
    return  [f"customer{n}" for n in random.choices(range(100),  k = 100)]

@task
def process_customer(customer_id: str) -> str:
    # Process a single customer
    return f"Processed {customer_id}"

@flow
def main() -> list[str]:
    customer_ids = get_customer_ids()
    # Map the process_customer task across all customer IDs
    results = process_customer.map(customer_ids)
    return results

if __name__ == "__main__":
    main.serve(name = "My-First-Deployment")
```
